```python
from citipy import citipy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from config import key_dict
from pprint import pprint
import scipy.stats as sts
```

```python


weather_df = pd.DataFrame(columns = ["Lat","Long","City","Country", 'Temp(F)', 'Humidity(%)', "Cloudiness(%)",
                                                                     "Wind Speed(mph)"])

key = key_dict['owm']
base_url = "https://api.openweathermap.org/data/2.5/weather?"
units = "imperial"


while len(weather_df) < 500:
    lat = np.random.uniform(low = -90, high = 90)
    lng = np.random.uniform(low = -180, high = 180)
    city_name = citipy.nearest_city(lat,lng).city_name.title()
    
    #is unique city?
    if city_name in list(weather_df['City']):
        print(city_name, "was not a unique city. Trying new coordinates.")
        continue
    
    #great, get country code, construct URL, search for weather data
    country_code = citipy.nearest_city(lat,lng).country_code.upper()
    
    url = str(f"{base_url}q={city_name},{country_code}&appid={key}&units={units}")    
    weather_data = requests.get(url).json()
    
    #try for weather data from city; does it exist in weather api?
    if weather_data['cod'] != 200:
        print(city_name, "did not have weather data. Trying new coordinates")
        #if there's no weather data, we need a new city, so reloop and randomly generate coords
        continue
    
    #if we make it this far, attach all the info
    city_information = {
        #note: the coordinates of the city itself might actually be wildly different 
        #than the coordinates originally searched for, so I'm updating these values in the DF
        'Lat':weather_data['coord']['lat'],
        'Long':weather_data['coord']['lon'],
        'City':city_name,
        'Country':country_code,
        'Temp(F)':weather_data['main']['temp'],
        'Humidity(%)':weather_data['main']['humidity'], 
        "Cloudiness(%)":weather_data['clouds']['all'], 
        "Wind Speed(mph)":weather_data['wind']['speed']
    }
    
                                
    weather_df = weather_df.append(city_information,
                                   ignore_index = True)
    print('Adding',city_name,'information to dataframe -- city number', len(weather_df))
```

```python
northern_hemi = weather_df.loc[weather_df['Lat'] >= 0,:]

slope, intercept, r, p, se = sts.linregress(northern_hemi['Temp(F)'], northern_hemi['Lat'])
reg_eqn = f"Y = {round(slope, 3)}X + {round(intercept,3)}"

plt.scatter('Temp(F)', 'Lat', data = northern_hemi, edgecolor = 'black')
plt.plot(northern_hemi['Temp(F)'], northern_hemi['Temp(F)']*slope + intercept, color = 'red')
plt.annotate(xy = (-5,18), s = (reg_eqn), color = 'red', size = 14)
plt.ylabel('Latitude degrees')
plt.xlabel('Temp in F')
plt.title('Comparing Tempterature to Latitude\n in the Northern Hemishpere')
plt.grid()
plt.show()
plt.savefig(fname = 'Resources/nothern_temperature.png',format = 'png')
```

The above graph, a graph comparing latitude to temperature in the northern hemisphere, shows us that as one gets closer to the equator, the temperature rises.
